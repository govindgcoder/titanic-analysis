```python
#import required libraries
import numpy as np
import pandas as pd

#import dataset
titanic = pd.read_csv("Titanic-Dataset.csv")
titanic.head()
```

```python
#missing values
titanic.isna()
```

```python

#dropping missing columns
titanic = titanic.dropna()
titanic.head()

#let's do some EDA
#basic bar graph using matplot - relation between survived and passenger class
titanicS = titanic[titanic["Survived"]==1]
relSandPc = titanicS["Pclass"].value_counts()
print(relSandPc)
```

```python

#using matplotlib
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x=[0,1,2]
y=relSandPc.to_numpy()
print(x, y)
ax.bar(x,y,color=["#36F86F"])
ax.set_ylabel('Survivors')
ax.set_title('Class')

ax.legend(title='Relationship between Passenger Classes and Survivors',labels=["x","y"])
plt.show()
```

```python
#new feature: familySize
familySize = titanic["SibSp"]+titanic["Parch"]
titanic["familySize"]=familySize
titanic.head(15)
```

```python
#more analysis: with seaborn
import seaborn as sns
sns.barplot(data=titanic, x="Sex", y="Survived")
```

```python
#So, its clear females had a higher chance of survival
#What about ages?
relAandS = titanicS["Age"].value_counts()
relAandSdf = pd.DataFrame({"index": relAandS.index, "Survivors": relAandS})

sns.lineplot(data=relAandSdf, x="index", y="Survivors")
```

```python
#another view
sns.kdeplot(data=relAandSdf, x="index", y="Survivors")
```

```python
#data is condensed around the middle age group of 25-40, with most of the survivors
#now just to look at correlation between different features
#for that, lets select numerical features first
titanicN = titanic.select_dtypes(include=["number"])
sns.heatmap(titanicN.corr(), cmap="coolwarm", annot=True)
```

```python
#from figure, its clear that parents and siblings have moderate positive linear correlation, which points to families
#Fare and passenger class have a moderate-to-strong negative linear correlation, which makes sense since class 1 is costlier than class 3
#Passenger class & survival - negative correlation - shows that richer classes had more survival chances
#Age and Passenger class - negative correlation - younger people favored higher classes
#Weak negative correlation between SibSp and Age suggests younger passengers had their siblings with them.
#There is a very weak correlation between Fare and Survival suggesting the Cost of the ticket had played a minor role, 
#supporting the reasoning of Survival of Higher Classes.
```
